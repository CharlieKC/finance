```python
# Import yfinance package
import yfinance as yf
from datetime import date

# Set the start and end date
start_date = '1990-01-01'
today = str(date.today())
# Set the ticker
end_date = today

ticker = 'MMAT'

# Get the data
data = yf.download(ticker, start_date, end_date)

# Print 5 rows
data.tail()
```
